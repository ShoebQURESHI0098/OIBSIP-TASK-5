```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample DataFrame (Replace with actual data)
np.random.seed(42)
df = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Sales': np.random.randint(100, 500, 5)
})

# Define colors using seaborn palette
colors = sns.color_palette("pastel")

# Create the pie chart
plt.figure(figsize=(7, 7))
plt.pie(df['Sales'], labels=df['Category'], autopct='%1.1f%%', colors=colors, startangle=140, wedgeprops={'edgecolor': 'black'})

# Title
plt.title('Sales Distribution by Category', fontsize=14, fontweight='bold')

# Show plot
plt.show()
```

```python
                                                       #Model Building#
```

```python
X = df.loc[:, df.columns != 'Sales']  # Select all columns except 'Sales'
y = df.loc[:, 'Sales']  # Select only the 'Sales' column
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

def evaluate_model(y_true, y_pred):
    metrics = {
        "MAE": mean_absolute_error(y_true, y_pred),
        "MSE": mean_squared_error(y_true, y_pred),
        "RMSE": np.sqrt(mean_squared_error(y_true, y_pred)),
        "RÂ² Score": r2_score(y_true, y_pred)
    }
    return metrics
```

```python
                                                  #Linear Regression Model
```

```python
from sklearn.linear_model import LinearRegression

model_name = "LinearRegression"

# Build a Linear Regression model
lin_reg = LinearRegression()

# Fit the model on the training data
lin_reg.fit(X_train, y_train)

# Predict using the model on the data
predictions = lin_reg.predict(X_test)

# Function call to Evaluate results of 'MAE', 'MSE', 'RMSE', 'R-squared'
evaluation_results_l = evaluate_model(y_test, predictions) # type: ignore

evaluation_results_l["Model"] = model_name


models = pd.DataFrame([evaluation_results_l])

print("Evaluation Results for", model_name)
print('-'*30)
for metric, value in evaluation_results_l.items():
    print(f"{metric}: {value}")
```

```python
                                             #Randome Forest model
```

```python
from sklearn.ensemble import RandomForestRegressor
import pandas as pd

# Ensure the evaluation function is defined
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

def evaluate_model(y_true, y_pred):
    metrics = {
        "MAE": mean_absolute_error(y_true, y_pred),
        "MSE": mean_squared_error(y_true, y_pred),
        "RMSE": np.sqrt(mean_squared_error(y_true, y_pred)),
        "RÂ² Score": r2_score(y_true, y_pred)
    }
    return metrics

# Train Random Forest Model
rf_model = RandomForestRegressor(n_estimators=100, random_state=0)
rf_model.fit(X_train, y_train)

# Make predictions
y_pred = rf_model.predict(X_test)

# Evaluate model
evaluation_results_r = evaluate_model(y_test, y_pred)  # Use correct function name

# Add model name
evaluation_results_r["Model"] = "Random Forest"

# Store results in DataFrame
models = pd.DataFrame([evaluation_results_r])

# Display results
print(f"Evaluation Results for {evaluation_results_r['Model']}")
print("-" * 40)
for metric, value in evaluation_results_r.items():
    if metric != "Model":
        print(f"{metric}: {value:.4f}")
    else:
        print(f"{metric}: {value}")
```

```python
#Gradient Boosting Regressor MODEL 
```

```python
from sklearn.ensemble import GradientBoostingRegressor
import pandas as pd

# Train Gradient Boosting Model
gb_model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=0)
gb_model.fit(X_train, y_train)

# Make predictions
y_pred = gb_model.predict(X_test)

# Evaluate model
evaluation_results_gb = evaluate_model(y_test, y_pred)  # Use correct function

# Add model name
evaluation_results_gb["Model"] = "Gradient Boosting"

# Convert results to DataFrame
evaluation_df = pd.DataFrame([evaluation_results_gb])

# Display results
print(f"Evaluation Results for {evaluation_results_gb['Model']}")
print("-" * 40)
for metric, value in evaluation_results_gb.items():
    if isinstance(value, (int, float)):  # Check if the value is numeric before formatting
        print(f"{metric}: {value:.4f}")
    else:
        print(f"{metric}: {value}")  # Print string values without formatting
```

```python
                                  #Decision Tree Regressor Model
```

```python
from sklearn.tree import DecisionTreeRegressor
import pandas as pd

# Initialize Decision Tree Regressor
dt_model = DecisionTreeRegressor(max_depth=5, random_state=0)

# Train the model
dt_model.fit(X_train, y_train)

# Make predictions
y_pred = dt_model.predict(X_test)

# Evaluate model performance
evaluation_results_dt = evaluate_model(y_test, y_pred)  # Assuming evaluate_model() function is defined

# Add model name
evaluation_results_dt["Model"] = "Decision Tree Regressor"

# Store results in DataFrame
evaluation_df = pd.DataFrame([evaluation_results_dt])

# Display results
print(f"Evaluation Results for {evaluation_results_dt['Model']}")
print("-" * 40)
for metric, value in evaluation_results_dt.items():
    if isinstance(value, (int, float)):
        print(f"{metric}: {value:.4f}")
    else:
        print(f"{metric}: {value}")
```
